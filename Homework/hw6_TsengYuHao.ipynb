## Setting figure frame

```julia
x = Array(-10:10)
y1 = 2*x
y2 = x.^2
plot(x,y1)
figure()
plot(x,y2)
```

```julia
subplot(2,1,1)
plot(x,y1)
axhline(0,color = "gray")
axvline(0,color = "gray")

subplot(2,1,2)
plot(x,y2)
grid()
```

## Sharing the same x-axes

```julia
plot(x,y1,"b")
tick_params("y",colors = "b")

twinx() # sharing the same x-axes

plot(x,y2,"r")
tick_params("y",colors = "r",direction = "in")
```

```julia
plot(x,y1,"b")
tick_params("y",colors = "b")

twinx() # sharing the same x-axes

plot(x,y2,"r")
tick_params("y",colors = "r",direction = "in")
xlim(-10,0) # set x-axis from -10 to 0
ylim(0,100) # set x-axis from -10 to 0
```

## "xlim" and "ylim" can also be used for reversing axes

```julia
plot(x,y1,"b")
tick_params("y",colors = "b")

twinx() # sharing the same x-axes

plot(x,y2,"r")
tick_params("y",colors = "r",direction = "in")
xlim(0,-10) # reverse x-axis
ylim(100,0) # reverse y-axis
```

## Labels and annotations
### Showing legends of curves/lines

```julia
plot(x,y1,"b",label="y=2x")
plot(x,y2,"r",label="y=x^2")
legend(loc="upper right")
```

## Showing text in a figure and changing the text color

```julia
plot(x,y1,"b",label="y=2x")
plot(x,y2,"r",label="y=x^2")
legend(loc="upper right")

annotate("minimum point",xy=(2.5,-10),color="r")
```

## Showing an arrow

```julia
plot(x,y1,"b",label="y=2x")
plot(x,y2,"r",label="y=x^2")
legend(loc="upper right")

annotate("minimum point", xytext=(2.5,-10), color="r", xy=(0,0), arrowprops = Dict("facecolor"=>"r"))
```

# Plotting styles

```julia
plot(x,y1,"b,-",label="y=2x")
plot(x,y2,"rs--",label="y=x^2")
legend(loc="upper right")
```

## fills with colors between two curves/lines

```julia
plot(x,y1,"b")
plot(x,y2,"r")
fill_between(x, 0, y2, facecolor = "r") # Fill red between y = 0 and y = y2
```

```julia
plot(x,y1,"b")
plot(x,y2,"r")
fill_between(x, 0, y2, facecolor = "r") # Fill red between y = 0 and y = y2
fill_between(x, 0, y1, facecolor = "b") # Fill blue between y = 0 and y = y1
```

## Adjusting transparency

```julia
plot(x,y1,"b")
plot(x,y2,"r")
fill_between(x, 0, y2, facecolor = "r",alpha = 0.5) 
fill_between(x, 0, y1, facecolor = "b",alpha = 0.5) 
```

## Plot the curve A given as “A=cos(2π t/100)”, where t is from 1~500 seconds. Show curve A by a “dashed” line. Shade the areas from zero to positive A by red color. Show the axes of x and y as well and label the name for the x-y axes.
 


```julia
t = 1:0.1:500
A = cos.(2*pi*t/100)
plot(t,A,"k--")
axhline(0,color = "gray")
axvline(0,color = "gray")
# A_positive = A
# A_positive[findall(A_positive.<0)] .= 0
# fill_between(t, 0, A_positive, facecolor="r")
fill_between(t[findall(A.>=0)], 0, A[findall(A.>=0)], facecolor="r")
xlabel("time (seconds)")
ylabel("Amplitude")
```
