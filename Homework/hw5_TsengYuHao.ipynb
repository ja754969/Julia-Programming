## For the decaying of $x=1000, 1000/2, 500/2, 250/2, 125/2, …,$ find out the length of x until x is just smaller than $10^{-8}$. 
### (HINT1: use “while loop”)  
### (HINT2: set $x=1000$ first, then put in the new value to x, like: $x=[x;500]$, then we will have $x=[1000; 500]$)  
### (HINT3: coding for $10^{-8}$ : ```10.0^(-8)```)  
__________________________________________



````julia
# x = Array([])
# concat(x,Array([]))
x = 1000
length_of_x = 0
while x >= 10^(-8)
    x = x/2
    length_of_x = length_of_x + 1
end
println(length_of_x)
````

## (Bonus: 總分+1%) Use a "while loop" to find the factorial (階乘) of a given integer.

````julia
given_int = 5
factorial_given_int = given_int 
while given_int != 0
    factorial_step = given_int
    if factorial_given_int == given_int
        factorial_given_int = given_int
    else
        factorial_given_int = factorial_given_int*factorial_step 
    end
    given_int = given_int-1
end
println(factorial_given_int)
````

## (70%) Assuming we have 100 persons in a class, everyone rolls a dice until he/she gets the number “6”. We use Julia Language to roll the dice for 100 persons. Record the number of times when he/she gets the number “6”, then use a plot to show how many persons get “6” at the first time, and how many persons get “6” at the second time, and so on. 

````julia
# rand(1:6)

# Array{Int64}(zeros(10))
function dice_6_distribution(n_people)
    n_count = Array{Int64}(zeros(n_people)) # The array of different person
    for i = 1:1:n_people
        while rand(1:6)!=6
            n_count[i] = n_count[i]+1
        end
        n_count[i] = n_count[i]+1
    end
    n_count_result = n_count
    return n_count_result
end
````

````julia
n_count = dice_6_distribution(100)
# println(n_count)
dice_6_count_plot = Array{Int64}(zeros(maximum(n_count))) # The number of people relative to their dice times 
for i = 1:maximum(n_count)
    find_count = findall(n_count .== i) # Find the number of times from minimum to maximum
    dice_6_count_plot[i] = length(find_count) # How many people that belong to this number of times
end
# println(dice_6_count_plot) # Check the answer
plot(1:maximum(n_count),dice_6_count_plot,".-b",label = "number of people")
xlabel("times")
ylabel("number of people")
title("Figure of dice 6 distribution")
legend()
grid()
````
